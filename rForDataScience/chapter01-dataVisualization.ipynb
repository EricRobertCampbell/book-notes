```r
?geom_histogram
```

`bins` sets the **number** of bins to use.

4. Make a histogram of the `carat` variable in the `diamonds` dataset. What value of `binwidth` reveals the most interesting patterns?

```r
for (bin in c(5, 10, 15, 20, 30, 40, 50, 100)) {
    p <- ggplot(diamonds, aes(carat)) +
        geom_histogram(bins = bin) +
        labs(title = paste(bin))
    print(p)
}
```

40 or 50 seem to show the most interesting pattern - you can see the various spikes, but there's not so many that it becomes difficult to see the forest for the trees (like with 100 bins).

## Visualizing Relationships

### A numerical and categorical variable

```r
ggplot(penguins, aes(species, body_mass_g)) +
    geom_boxplot()
```

```r
ggplot(penguins, aes(body_mass_g, colour = species, fill = species)) +
    geom_density(linewidth = 0.75, alpha = 0.5)
```

```r
ggplot(penguins, aes(island, fill = species)) +
    geom_bar()
```

```r
ggplot(penguins, aes(island, fill = species)) +
    geom_bar(position = 'fill')
```

```r
ggplot(penguins, aes(flipper_length_mm, body_mass_g)) +
    geom_point(aes(colour = species, shape = species)) +
    facet_wrap(~island)
```
